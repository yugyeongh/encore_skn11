### Figure | Axes

- Figure: 전체 그림이 그려지는 컨테이너
- Axes: 그래프가 실제로 그려지는 영역 (좌표, 데이터)

```python
fig = plt.figure(figsize=(6,4), facecolor='green')
print(fig)

ax = plt.gca()
ax.plot([1,2,3],[4,5,6])
print(ax)

plt.title('Hello Matplotlib')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### plt 속성 설정

```python
plt.plot(x,y)
plt.title('제목')
plt.xlabel('시간')
plt.ylabel('점수')
plt.show()
```

```python
# ticks (눈금)
plt.figure(figsize=(3,4))
x = np.arange(1,101)
y = np.linspace(2, 200, 100)

plt.xticks(ticks=np.arange(0,101,5), rotation=90)
plt.yticks(rotation=45) # ticks: 눈금 기울기 설정

plt.plot(x,y)
plt.show()
```

```python
# lim (보여지는 범위 재지정)
plt.figure(figsize=(3,4))
x = np.arange(1,101)
y = np.linspace(2, 200, 100)

plt.xlim(1, 50)
plt.ylim(2, 100)

plt.plot(x,y)
plt.show()
```

```python
# legend (범례)
plt.figure(figsize=(3,4))
x = np.arange(1,101)
y = np.linspace(2, 200, 100)

x = np.arange(1,101)
y = np.linspace(2, 200, 100)
y2 = np.linspace(3, 300, 100)

plt.plot(x,y, label='시간')
plt.plot(x, y2, label='공간')

# loc 범례 위치 (default: best(알아서 최적의 위치를 찾아줌))
# - (upper | lower) (left | right)
# - center
plt.legend(loc='center') # loc: 범례의 위치를 지정해주는 속성

# plt.plot(x,y) # 이게 있으면 왜 그래프 색이 달라지지
plt.show()
```

### 그래프 여러 개 그리기

```python
# 여러 종류의 그래프를 동시에 그리기
plt.figure(figsize=(3,4))
x = np.arange(1,101)
y = np.linspace(2, 200, 100)

x = np.arange(1,101)
y = np.linspace(2, 200, 100)
y2 = np.linspace(3, 300, 100)

plt.plot(x, y, label='선', color='red', marker='o')
plt.bar(x,y2, label='바', color='green')

plt.legend() # label 지정
plt.grid(True)

plt.show()
```

```python
# subplots를 이용해서 여러 그래프를 각각 그리기
fig, ax = plt.subplots(1,2, figsize=(10,5))

x1 = np.arange(1,11)
y1 = x1*2
x2 = np.arange(11,20)
y2 = x2*2

ax[0].plot(x1,y1, label='선', color='red')
ax[0].set_xlabel('X1') # set이 붙으면 좀 더 객체 지향적임
ax[0].set_ylabel('Y1')
ax[0].set_title('빨간 선')
ax[0].legend()
ax[0].grid()

ax[1].bar(x2,y2, label='바', color='yellowgreen')
ax[1].set_xlabel('X2') # set이 붙으면 좀 더 객체 지향적임
ax[1].set_ylabel('Y2')
ax[1].set_title('연두 막대 그래프')
ax[1].legend()
ax[1].grid()

plt.show()
```
