### 통계 관련 numpy

```python
import numpy as np
```

##### 중심 경향 측도

```python
height = np.array([160,170,180,190,200,150,165,175,176,178,168,182,183,157,160])
```

```python
# 평균값값
print(np.mean(height))
```

```python
# 중위값
print(np.median(height))
```

```python
# 최빈값
values, count = np.unique(height, return_counts=True)
print(values, count)

max_count = np.max(count)
print(values[count == max_count])
```

# 산포 측도

```python
# 분산 variance
# (값 - 평균)^2의 평균

var1 = np.mean(np.square(height - np.mean(height)))
var2 = np.var(height) # 분산 구하는 함수
print(var1, var2)
```

```python
# 표준편차 standard deviation
# 분산의 제곱근

std1 = np.sqrt(var1)
std2 = np.std(height) 
print(std1, std2)
```

```python
# 범위
print(np.max(height)-np.min(height))
```

### 확률분포

```python
import matplotlib.pyplot as plt
```

```python
# 정규분포
# np.random.randn(shape)
# 평균이 0, 표준편차가 1인 난수 출력 함수

np.random.randn()
```

```python
arr = np.random.randn(1000)

plt.hist(arr, bins=100) # 히스토그램, bins로 막대개수 지정하기
plt.show() # 출력
```

```python
# np.random.normal(loc,scale,size)
# - loc: 평균
# - scale: 표준편차
# - size: 갯수 or normal
# arr = np.random.normal(10,1,1000)
arr = np.random.normal(10,1,(10,100)) #shape주기
print(arr)

plt.hist(arr, bins=10)
plt.show()
```

```python
# 균등 분포포
# np.random.rand(num)
# - 0.0 이상 ~ 1.0 미만 수에서 균등하게 난수 추출
# - num: 갯수 or shape
arr = np.random.rand(1000)
arr = np.random.rand(2,3)

plt.hist(arr)
plt.show()
```

```python
# 균등분포
# np.random.randint(start, end, size)
# - start 이상 end 미만의 난수(정수)를 size만큼 추출해준다.

np.random.seed(0) # 난수계산을 시작하는 seed값이 고정되어 매번 같은 수가 추출됨

arr = np.random.randint(1,100,1000)
print(arr)

plt.hist(arr)
plt.show()
```
