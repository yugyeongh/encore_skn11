# 활성화 함수 (Activation Function)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-5, 5, 100)
```

---

### 계단 함수 (Step Function)

$ f(x) = \begin{cases} 1, &x \geq 0 \\ 0, & x <0 \end{cases} $

- 입력값이 0이상이면 1, 그렇지 않으면 0 출력
- 미분 불가하고, 작은 변화에도 값이 급격히 변하여 역전파 학습에 부적합

```python
def step_function(x):
    return np.where(x >= 0, 1, 0)
```

```python
plt.figure(figsize=(5,3))
plt.plot(X, step_function(X))
plt.title('Step Function')
plt.grid()
plt.ylim(-0.1, 1.1)
plt.show()
```

---

### 시그모이드 함수 (Sigmoid Function)

$ f(x) = \frac{1}{1 + e^{-x}} $

- 출력범위: (0,1)
- 매끄러운 S자 곡선 형태로 확률 해석 가능
- 기을기 소실(Vanishing Gradient) 문제 발생 가능

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.figure(figsize=(5,3))
plt.plot(X, sigmoid(X))
plt.title('Sigmoid Function')
plt.grid()
plt.ylim(-0.1, 1.1)
plt.show()
```

---

### 하이퍼볼릭 텐젠트 함수 (Hyperbolic Tangent, Tanh)

$ f(x) = \frac{e^7 - e^{-x}}{e^x + e^{-x}} $

- 출력 범위: (-1,1)
- 시그모이드보다 중심이 0에 가까워 더 빠른 학습 진행 가능
- (시그모이드를 변형시킨거기 때문에 시그모이드와 마찬가지로) 기울기 소실 문제 발생 가능 

```python
def tanh(x):
    #return (np.exp(x) - np.exp(-x) / np.exp(x) + np.exp(-x))
    return np.tanh(x)
```

```python
plt.figure(figsize=(5,3))
plt.plot(X, tanh(X))
plt.title('Tanh Function')
plt.grid()
plt.ylim(-1.1, 1.1)
plt.show()
```

---

### ReLU (Rectified Linear Unit)

$ f(x) = \max(0,x) $

- 출력 범위: $ (0, \infty) $
- 음수 입력에 대해 0을 출력하여 계산이 간단하고 학습 속도가 빠름
- 죽은 뉴런(Dying ReLU) 문제 발생 가능

```python
def relu(x):
    return np.maximum(0,x)
```

```python
plt.figure(figsize=(5,3))
plt.plot(X, relu(X))
plt.title('ReLU Function')
plt.grid()
plt.show()
```

---

### Leaky ReLU

$ f(x) = \begin{cases} x, & x \geq 0 \\ \alpha x, & x < 0\end{cases} $

- 출력 범위: $ (-\infty, \infty) $
- ReLU의 죽은 뉴런 문제를 해결하기 위해 음수 입력에 작은 기울기 $\alpha$를 적용 ($\alpha$는 일반적으로 0.01 사용)

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x>=0, x, alpha*x)
```

```python
plt.figure(figsize=(5,3))
plt.plot(X, leaky_relu(X))
plt.title('Leaky ReLU Function')
plt.grid()
plt.show()

# 음수에서 미세하게 기울기가 있는 것을 볼 수 있음 -> bc 기본값으로 0.01을 줬기 때문
```

```python
plt.figure(figsize=(5,3))
plt.plot(X, leaky_relu(X, alpha=0.1))
plt.title('Leaky ReLU Function')
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd

np.random.seed(0)
X = np.random.uniform(-5, 5, size=10) # -5에서 5 사이에 균등분포를 따르는 10개의 랜덤한 숫자

df = pd.DataFrame({
    "input (x)": X,
    "Step Function": step_function(X),
    "Sigmoid": sigmoid(X),
    "Tanh": tanh(X),
    "ReLU": relu(X),
    "Leaky ReLU": leaky_relu(X)
})
df
```

```python
plt.figure(figsize=(8,6))
plt.plot(df['input (x)'], df['Step Function'], marker='o', label='step func')
plt.plot(df['input (x)'], df['Sigmoid'], marker='o', label='sigmoid func')
plt.plot(df['input (x)'], df['Tanh'], marker='o', label='tanh')
plt.plot(df['input (x)'], df['ReLU'], marker='o', label='ReLU')
plt.plot(df['input (x)'], df['Leaky ReLU'], marker='o', label='Leaky ReLU')

plt.xlabel('input (X)')
plt.ylabel('Activation Output')
plt.legend()
plt.grid()
plt.show()
```
