```python
import yt_dlp

url = 'https://youtu.be/wBLsojzEpRA?si=66lULGyz7HruI9rU'
output_path = 'youtube_output'

ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl': output_path,
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }]
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])

print('youtube_output.mp3 다운로드 완료')
```

```python
import whisper

model = whisper.load_model('small')
result = model.transcribe('youtube_output.mp3')
print(result['text'])
```
