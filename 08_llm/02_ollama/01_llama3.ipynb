### 3. 파라미터 조절

```python
import ollama
import time

def generate_text_with_params(prompt, max_length=50, temperature=0.7, top_p=1):
    print(f'[Prompt: {prompt} | max_length({max_length}), temperature({temperature}), top_p({top_p})]')

    start_time = time.time()
    response = ollama.generate(
        model = 'llama3',
        prompt = prompt,
        options = {
            'max_length': max_length,
            'temperature': temperature,
            'top_p': top_p
        }
    )
    end_time = time.time()

    print(response['response'])
    print('소요 시간:',end_time - start_time)

    return response['response']
```

```python
prompt = '오늘 저녁 메뉴를 추천해줘 나는 김밥이 먹고 싶어서 김밥 종류를 추천해주면 좋겠어 그리고 어울리는 컵라면도 같이 알려줘'

max_lengths = [20,50,100]
temperatures = [0.3, 1.2]

for max_length in max_lengths:
    for temperature in temperatures:
        text = generate_text_with_params(prompt, max_length, temperature, top_p=1)
        print(text)
        print('='*100)
```

### 4. 응답에 특정 단어를 포함하도록 하기

```python
def generate_text_with_word(prompt, required_word):
    print(f'[ Prompt: {prompt} ] | 포함할 단어: {required_word}]')
    
    prompt = prompt + f"<Please make sure to include the word '{required_word}' in your response.>"
    
    response = ollama.generate(model='llama3', prompt=prompt)

    res_txt = response['response']

    if required_word.lower() in res_txt.lower():
        print('✅ 성공')
    else:
        print('🚨 실패')

    return res_txt
```

```python
test_cases = [
    ('Desribe the future of AI', 'truck'),
    ('Tell me about space exploration','Mars'),
]

for prompt, required_word in test_cases:
    response = generate_text_with_word(prompt, required_word)
    print(f'Response: {response}')
    print('='*100)
```

### 5. 다국어 처리 실험

```python
def generate_text_in_language(language, prompt):
    print(f'[ Language: {language} | Prompt: {prompt}]')
    
    response = ollama.generate(model='llama3', prompt=prompt)
    print(response['response'])
    print('='*100)
```

```python
prompts = {
    "English":"Please recommend a dinner menu for me today. Please answer in English",
    "한국어":"오늘 저녁 메뉴 1가지만 추천해줘. 한국어로 대답해줘",
    "日本語(にほんご)":"今日は夕食のメニューをおすすめしてください。日本語で答えてください。",
    "汉语(简体字)":"请推荐今天的晚餐菜单。请用中文回答。",
    "français":"Veuillez me recommander un menu pour le dîner aujourd'hui."
}

for language, prompt in prompts.items():
    generate_text_in_language(language, prompt)
```

### LLaMA 모델을 이용한 간단한 대화 시스템

- 사용자 입력 input()
- 응답 print()로 출력
- 'exit' 입력 시 대화 종료

```python
while True:
    prompt = input('대화를 입력하세요:\n')

    if prompt.lower() == 'exit':
        break

    korean_prompt = f"{prompt} 나는 영어를 하나도 모르는 사람이라 영어로 답변해주면 이해하지 못해 그러니까 영어로 나온 답변은 꼭 한글로 번역해서 대답해주고, 한글 답변만 보여주고 한글 답변은 보여주지마 "
    response = ollama.generate(model='llama3', prompt=korean_prompt)
    
    print(response['response'])
    print('='*100)
```
