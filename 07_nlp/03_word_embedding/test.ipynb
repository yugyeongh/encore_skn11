### 자연어 딥러닝 기초
> 신경망 모델을 활용해 의미를 학습하고 작업을 수행하는 방법론

- 자연어 딥러닝은 시퀀스 모델에 기반함
- 시퀀스 모델: 연속적인 데이터를 다룰 수 있게 만든 모델
<br/>


- input으로 넣어준 x가 시퀀셜 데이터
- 시퀀셜 데이터는 순서가 중요
- 순차적으로 hidden layer에 들어감
- U,V,W : 가중치
- RNN은 시간의 흐름에 따라 변화하는 데이터를 처리하기 때문에
    - 이전 상태를 기억하고
- hidden state: 이전 시점의 정보를 현재 시점으로 전달
    - 즉, 이전 시퀀스의 정보를 현재 시점으로 전달하는 것
