### 텍스트 전처리

```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
spam_df = pd.read_csv('sms_spam.csv')
spam_df.head()
```

```python
corpus = spam_df['CN'][0]
corpus
```

```python
# 명사 추출
okt = Okt()
nouns = okt.nouns(corpus)
nouns
```

```python
# 명사 빈도수 세기
word_count = Counter(nouns)
word_count
```

```python
# 불용어 처리
ko_stopwords = ['은','는','이','기','및','더','수']

word_count = {word: count for word, count in word_count.items() if word not in ko_stopwords}
word_count
```

### WordCloud 객체 생성

```python
wordcloud = WordCloud(
    width=800,
    height=800,
    font_path='C:\\Windows\\Fonts\\malgunbd.TTF',
    background_color='white'
).generate_from_frequencies(word_count)     # generate_from_frequencies: 단어와 빈도수를 받아서 워크 클라우드를 생성하는 메서드

wordcloud       # 워드 클라우드 객체 만들기 성공
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

### 전체 corpus로 WordCloud 그리기!

```python
corpus = spam_df['CN']
corpus.info()
```

```python
# word_count = Counter(corpus)
text = ' '.join(corpus)

wordcloud = WordCloud(
    width=1000,
    height=1000,
    font_path='C:\\Windows\\Fonts\\malgunbd.TTF'
).generate(text)

# 단어 구름 출력
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 축 숨기기
plt.show()
```
